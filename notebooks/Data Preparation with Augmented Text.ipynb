```python
import pandas as pd
```

```python
df1 = pd.read_csv("severe_toxic_generated.csv")
df2 = pd.read_csv("threat_generated.csv")
df3 = pd.read_csv("identity_hate_generated.csv")
df = pd.read_csv("train.csv")
```

### Data Processing

```python
df1.drop("Unnamed: 0", axis=1, inplace=True)
df2.drop("Unnamed: 0", axis=1, inplace=True)
df3.drop("Unnamed: 0", axis=1, inplace=True)
df.drop("id", axis=1, inplace=True)
```

```python
df.rename(columns = {'comment_text':'text'}, inplace = True)
```

### Concatenating original training data with augmented data

```python
frames = [df, df1, df2, df3]
result = pd.concat(frames)
```

### Converting datatype of labels from float to integer

```python
result = result.astype({'toxic': 'int32', 'severe_toxic': 'int32', 'obscene': 'int32', 'threat': 'int32', 'insult': 'int32', 'identity_hate': 'int32'})
```

### Shuffling data to enable fair distribution of categories across training and validation batches

```python
result = result.sample(frac = 1)
```
